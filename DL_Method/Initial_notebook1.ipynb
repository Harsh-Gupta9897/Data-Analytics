```python
def predict_wicket_based(over_rem,Z0,L):
    return Z0*(1- np.exp((-L*over_rem)/Z0))
    
    
```

```python
def fun_to_minimize(params,runs_rem, wickets_rem, overs_rem,n):
    squared_errors =[]
    for i in range(n):
        pred = predict_wicket_based(overs_rem[i],params[wickets_rem[i]-1],params[10])
        squared_errors.append((pred- runs_rem[i])**2)
         
    return np.sum(squared_errors)
        
        
```

```python
from scipy.optimize import minimize
df = df[df['Wickets.in.Hand']!=0]
runs_rem = df['Runs.Remaining'].tolist()
wickets_rem =df['Wickets.in.Hand'].tolist()
overs_rem=df['Over.Remaining'].tolist()
n = len(overs_rem)
model = minimize(fun_to_minimize,params,args=(runs_rem, wickets_rem, overs_rem,n),method='L-BFGS-B')
model.x
```

```python
model4 = minimize(fun_to_minimize,params,args=(runs_rem, wickets_rem, overs_rem,n),method='Powell')
print(model4)
```

```python
model5 = minimize(fun_to_minimize,params,args=(runs_rem, wickets_rem, overs_rem,n),method='TNC')

total_mse =0
for i in range(1,11):
    u= df[df['Wickets.in.Hand']==i]['Over.Remaining']
    n = len(u)
    predictions=[]
    y_test = df[df['Wickets.in.Hand']==i]['Runs.Remaining']
    Z0 = model5.x[i-1]
    L = model5.x[10]
    print("Error in Wicket Data :",i)
    for item in u:
        predictions.append(predict_wicket_based(item,Z0,L))
    m = metrics.mean_squared_error(y_test, predictions)
    total_mse = m+ total_mse
    print('MAE:', metrics.mean_absolute_error(y_test, predictions))
    print('MSE:', m)
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
    print()
print('Total_Mean_Squared_Error: ',total_mse)
```
